# Compare Cosine_similarity, Word_Vector and BLEU

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from sentence_transformers import SentenceTransformer
from pythainlp.tokenize import word_tokenize
from nltk.translate.bleu_score import sentence_bleu
from pythainlp.word_vector import WordVector

# Load the model for semantic similarity
model = SentenceTransformer('all-MiniLM-L6-v2')  # SentenceTransformer for semantic similarity,  'sentence-transformers/all-mpnet-base-v2'
wv = WordVector('thai2fit_wv') # corpus large Thai Word2Vec.  ltw2v_v1.0_15_window

# Cosine similarity using sentence embeddings from SentenceTransformer
def similarity_check(text1, text2):
  # SKLearn Cosine similarity
    embeddings = model.encode([text1, text2])
    cossim = cosine_similarity([embeddings[0]], [embeddings[1]])[0][0]
    cossim = min(cossim, 1.0)

  # bleu_score
    reference = [word_tokenize(text1)]
    candidate = word_tokenize(text2)
    Bleu_score = sentence_bleu(reference, candidate)

   # word_vectors
    vector_score = wv.similarity(text1, text2)


    print(f"Cosine similarity: {cossim}")
    print(f"BLEU score: {Bleu_score}")
    print(f"Word vector similarity: {vector_score}")

    return cossim, Bleu_score, vector_score
```

# Evaluation

**Function for Evaluating**

```python
import ast
import torch
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModel

# Load the model for semantic similarity
model = SentenceTransformer('all-MiniLM-L6-v2')  # SentenceTransformer for semantic similarity 'all-MiniLM-L6-v2', 'paraphrase-MiniLM-L6-v2'ม 'mrp/simcse-model-m-bert-thai-cased'

#WangchanBERTa
#model_name = "airesearch/wangchanberta-base-att-spm-uncased"
#tokenizer = AutoTokenizer.from_pretrained(model_name)
#model = AutoModel.from_pretrained(model_name)


# Function to calculate similarity between two texts using sentence embeddings
def calculate_similarity(text1, text2, threshold=0.9):  # Increased threshold for stricter matching
#WangchanBerTa need to handle pooling manually
   # inputs1 = tokenizer(text1, return_tensors="pt", truncation=True, padding=True, max_length=128)
   # inputs2 = tokenizer(text2, return_tensors="pt", truncation=True, padding=True, max_length=128)

    # Compute embeddings
   # with torch.no_grad():
   #     outputs1 = model(**inputs1)
   #     outputs2 = model(**inputs2)

    # Use mean pooling to compute sentence embeddings
   # embedding1 = outputs1.last_hidden_state.mean(dim=1)  # Average over token embeddings
   # embedding2 = outputs2.last_hidden_state.mean(dim=1)

    # Convert embeddings to numpy for cosine similarity calculation
   # embedding1 = embedding1.cpu().numpy()
   # embedding2 = embedding2.cpu().numpy()

    embeddings = model.encode([text1, text2])   #use this one in case of SentenceTransformer
    similarity_score = cosine_similarity([embeddings[0]], [embeddings[1]])[0][0]
    #similarity_score = cosine_similarity(embedding1, embedding2)
    return similarity_score

# Function to safely evaluate strings to lists (with format checking)
def safe_eval(data):
    """
    Safely parse the string as a Python literal (list) and check its format.
    """
    try:
        evaluated_data = ast.literal_eval(data)
        if is_valid_format(evaluated_data):
            return evaluated_data
        else:
            return []  # Return None if it's not valid
    except (ValueError, SyntaxError):
        return []  # Return None if it can't be evaluated

# Function to check if the data is in the correct format
def is_valid_format(data):
    return isinstance(data, list) and all(isinstance(item, list) and len(item) == 3 for item in data)

# Function to find the best matching aspect from the model output
def find_best_matching_aspect(true_aspect, model_output, threshold=0.9):
    best_match = None
    highest_similarity = threshold  # Set the threshold as the minimum similarity
    for model_triplet in model_output:
        if len(model_triplet) == 3:
            similarity = calculate_similarity(model_triplet[0], true_aspect)
            if similarity > highest_similarity:
                highest_similarity = similarity
                best_match = model_triplet
    return best_match

# Function to check matches for each comment
def check_comment_matches(ground_truth, model_output, threshold=0.9):
    matched_aspects = 0
    matched_sentiments = 0
    matched_opinion_words = 0

    remaining_model_output = model_output[:]  # Create a copy of model_output to track available matches

    for true_triplet in ground_truth:
        # Ensure the true_triplet has 3 elements
        if len(true_triplet) != 3:
            continue

        true_aspect = true_triplet[0]
        true_sentiment = true_triplet[1]
        true_opinions = true_triplet[2] if isinstance(true_triplet[2], list) else [true_triplet[2]]

        print(f"\nEvaluating ground truth aspect: {true_aspect}, sentiment: {true_sentiment}, opinions: {true_opinions}")

        # Find the best matching aspect in model output
        matching_triplet = find_best_matching_aspect(true_aspect, remaining_model_output, threshold)

        if matching_triplet:
            print(f"Matching model aspect found: {matching_triplet}")
            matched_aspects += 1

            # Check sentiment match
            if matching_triplet[1] == true_sentiment:
                matched_sentiments += 1
                print(f"Sentiment match for aspect '{true_aspect}'")

            # Check opinion word match
            model_opinions = matching_triplet[2] if isinstance(matching_triplet[2], list) else [matching_triplet[2]]
            for true_opinion in true_opinions:
                for model_opinion in model_opinions:
                    if calculate_similarity(model_opinion, true_opinion) > threshold:
                        matched_opinion_words += 1
                        print(f"Opinion word match: '{model_opinion}' matches '{true_opinion}'")
                        break

            # Remove the matched model aspect from remaining_model_output to prevent reuse
            remaining_model_output.remove(matching_triplet)

    return matched_aspects, matched_sentiments, matched_opinion_words

# Function to calculate precision, recall, and F1 score
def calculate_precision_recall_f1(matches, total_predicted, total_actual):
    precision = matches / total_predicted if total_predicted > 0 else 0
    recall = matches / total_actual if total_actual > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    return precision, recall, f1

# Function to calculate global scores for aspects, sentiments, and opinions
def calculate_global_scores(df, Label_column, model_column, threshold=0.9):  # Using higher threshold globally
    total_aspect_matches = 0
    total_sentiment_matches = 0
    total_opinion_matches = 0

    total_aspect_predicted = 0
    total_sentiment_predicted = 0
    total_opinion_predicted = 0

    total_aspect_actual = 0
    total_sentiment_actual = 0
    total_opinion_actual = 0

    for index, row in df.iterrows():
        # Safely parse the 'Label' and model output
        ground_truth = safe_eval(row[Label_column])
        model_output = safe_eval(row[model_column])

        # If parsing failed or model output is invalid, skip it
        if ground_truth is None or model_output is None or not is_valid_format(model_output):
            # Invalid output should not contribute to total_aspect_actual, but it should affect recall
            total_aspect_actual += len(ground_truth)  # Ground truth count still matters
            total_sentiment_actual += len(ground_truth)
            total_opinion_actual += sum(len(item[2]) if isinstance(item[2], list) else 1 for item in ground_truth)
            continue

        # Get the match scores for the current comment
        aspect_matches, sentiment_matches, opinion_matches = check_comment_matches(ground_truth, model_output, threshold)

        # Count matches
        total_aspect_matches += aspect_matches
        total_sentiment_matches += sentiment_matches
        total_opinion_matches += opinion_matches

        # Count total predicted and actual
        total_aspect_predicted += len(model_output)  # Total predicted aspects
        total_sentiment_predicted += len(model_output)  # Total predicted sentiments
        total_opinion_predicted += sum(len(item[2]) if isinstance(item[2], list) else 1 for item in model_output)  # Total predicted opinions

        total_aspect_actual += len(ground_truth)  # Total ground truth aspects
        total_sentiment_actual += len(ground_truth)  # Total ground truth sentiments
        total_opinion_actual += sum(len(item[2]) if isinstance(item[2], list) else 1 for item in ground_truth)  # Total ground truth opinions

    # Calculate precision, recall, and F1 for aspects
    precision_aspect, recall_aspect, f1_aspect = calculate_precision_recall_f1(total_aspect_matches, total_aspect_predicted, total_aspect_actual)

    # Calculate precision, recall, and F1 for sentiments
    precision_sentiment, recall_sentiment, f1_sentiment = calculate_precision_recall_f1(total_sentiment_matches, total_sentiment_predicted, total_sentiment_actual)

    # Calculate precision, recall, and F1 for opinions
    precision_opinion, recall_opinion, f1_opinion = calculate_precision_recall_f1(total_opinion_matches, total_opinion_predicted, total_opinion_actual)

    # Print the results clearly for each term
    print(f"\n{model_column} Performance:")
    print(f"Aspect - Precision: {precision_aspect:.2f}, Recall: {recall_aspect:.2f}, F1: {f1_aspect:.2f}")
    print(f"Sentiment - Precision: {precision_sentiment:.2f}, Recall: {recall_sentiment:.2f}, F1: {f1_sentiment:.2f}")
    print(f"Opinion - Precision: {precision_opinion:.2f}, Recall: {recall_opinion:.2f}, F1: {f1_opinion:.2f}\n")

    return {
        'aspect': (precision_aspect, recall_aspect, f1_aspect),
        'sentiment': (precision_sentiment, recall_sentiment, f1_sentiment),
        'opinion': (precision_opinion, recall_opinion, f1_opinion)
    }

# Example usage with a DataFrame
import pandas as pd

# Example DataFrame containing 'Label' and model outputs for different models
data = {
    'Label': [
        '[["หน้าปัดเรือนไมล์", "pos", "ชอบที่สุด"], ["จอกลาง", "neg", "แสบตา"], ["ทรง", "pos", "สวย"], ["ระบบแอร์", "pos", "ดี"]]',
        '[["ไฟหน้า", "neg", "ไม่สว่าง"], ["เครื่องเสียง", "pos", "ชัดเจน"]]'
    ],
    'Another Model': [
        '[["หน้าปัดเรือนไมล์", "pos", "สวย"], ["ระบบแอร์", "pos", "ดี"], ["ทรง", "pos", ["สวย", "ดุ"]], ["จอ", "neg", "แสบตา"]]',
        'Output: [["หมา", "neg", "มืด"], ["ไฟท้าย", "pos", "ชัดเจน"]]'  # Invalid output
    ]
}
# Test function
data = pd.DataFrame(data)
calculate_global_scores(data, 'Label','Another Model', threshold=0.9)
```
