Load Model

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("aisingapore/sea-lion-7b-instruct", trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained("aisingapore/sea-lion-7b-instruct", trust_remote_code=True)
```

Test Prompt

```python
prompt_template = "### USER:\n{human_prompt}\n\n### RESPONSE:\n"
prompt = """ในฐานะผู้เชี่ยวชาญด้านการลงทุนหุ้น ขอให้อ่านหัวข้อข่าวหุ้นและบอกสิ่งที่ปรากฏจากหัวข้อข่าวทั้งหมด4ส่วน คือ สัญลักษณ์หุ้น แง่มุม ความคิดเห็น และความรู้สึก เพื่อให้ผู้อ่านทั่วไปเข้าใจ
ในประโยคที่กำหนดให้ ต้องการให้ทำการจำแนก
1.สัญลักษณ์หุ้น(SYMBOL):ชื่อย่อของหุ้นที่ปรากฏในข่าว
2.แง่มุม(ASPECT):แง่มุมของหุ้นที่ปรากฏในข่าว
3.ความคิดเห็น(OPINION):ความคิดเห็นต่อแง่มุมนั้น
4.ความรู้สึก(SENTIMENT):เป็นบวกหรือเป็นลบหรือเป็นกลางต่อแต่ละแง่มุม
โดยในหนึ่งข่าวอาจมีหลายแง่มุม
ตัวอย่างข้อมูลเข้า-ข้อมูลออก:
ข้อมูลเข้า:
'COM7 ร่วง 13% เซ่นกำไร Q3 หดตัวแตะ 611 ล้านบาท หลังค่าใช้จ่ายพุ่ง'
ข้อมูลออก:
สัญลักษณ์หุ้น:COM7 แง่มุมของหุ้นที่ปรากฏในข่าว:ราคาหุ้น ความคิดเห็นต่อแง่มุมนั้น:ร่วง และความรู้สึก:ลบ,
สัญลักษณ์หุ้น:COM7 แง่มุมของหุ้นที่ปรากฏในข่าว:กำไร ความคิดเห็นต่อแง่มุมนั้น:หดตัว และความรู้สึก:ลบ,
สัญลักษณ์หุ้น:COM7 แง่มุมของหุ้นที่ปรากฏในข่าว:ค่าใช้จ่าย ความคิดเห็นต่อแง่มุมนั้น:พุ่ง และความรู้สึก:ลบ
ข่าวที่ให้ทำคือ:OSP ลุ้นกำไร Q3 โตเด่น รับรู้รายได้ธุรกิจเครื่องดื่มพุ่ง"""
full_prompt = prompt_template.format(human_prompt=prompt)

tokens = tokenizer(full_prompt, return_tensors="pt")
output = model.generate(tokens["input_ids"], max_new_tokens=512, eos_token_id=tokenizer.eos_token_id)
print(tokenizer.decode(output[0], skip_special_tokens=True))
```
