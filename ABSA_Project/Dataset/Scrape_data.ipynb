```python
import csv
import requests
from bs4 import BeautifulSoup
```

# **Kaohoon.com**

```python
def scrape_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    data = []
    for mag_box_container in soup.find_all(class_="mag-box-container clearfix"):
        for date, title, a_tag in zip(mag_box_container.find_all("h6"), mag_box_container.find_all("h5"), mag_box_container.find_all("a")):
            date_text = date.get_text().strip()
            title_text = title.get_text().strip()
            link = a_tag.get("href")
            if len(title_text) > 5:
                data.append([date_text, link, title_text])
    return data

# Initial URL
base_url = "https://www.kaohoon.com/latest-news/page/{}"
num_pages = 5  # Change this to the desired number of pages

# Open CSV file in write mode and write the header
with open("data_kaohoon.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Date", "Link", "Title"])  # Write header

    # Loop through each page and scrape data
    for page_number in range(1, num_pages + 1):
        current_url = base_url.format(page_number)
        data_current_page = scrape_page(current_url)

        # Write data to CSV
        writer.writerows(data_current_page)

print("Data has been saved to data_kaohoon.csv")
```
